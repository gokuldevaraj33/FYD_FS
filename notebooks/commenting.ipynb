```python
def unzip_docx(input_path, temp_dir):
    import os, shutil, zipfile
    if os.path.exists(temp_dir):
        shutil.rmtree(temp_dir)
    os.makedirs(temp_dir)
    with zipfile.ZipFile(input_path, 'r') as zip_ref:
        zip_ref.extractall(temp_dir)

unzip_docx(input_path, temp_dir)
```

```python
def load_document_xml(temp_dir, nsmap):
    doc_path = os.path.join(temp_dir, "word", "document.xml")
    doc_tree = etree.parse(doc_path)
    doc_root = doc_tree.getroot()
    return doc_tree, doc_root, doc_path

doc_tree, doc_root, doc_path = load_document_xml(temp_dir, nsmap)
```

```python
def display_all_runs(doc_root, nsmap):
    """
    Prints all text runs in the document, grouped by paragraph.
    """
    paragraph_count = 0
    for para in doc_root.findall(".//w:p", namespaces=nsmap):
        paragraph_count += 1
        print(f"\nüìù Paragraph {paragraph_count}")
        run_count = 0
        for run in para.findall(".//w:r", namespaces=nsmap):
            t_elem = run.find(".//w:t", namespaces=nsmap)
            if t_elem is not None and t_elem.text:
                run_count += 1
                print(f"   ‚û§ Run {run_count}: \"{t_elem.text}\"")
        if run_count == 0:
            print("   (No text runs found)")
```

```python
def flatten_runs_to_single_run(doc_root, nsmap):
    for para_idx, para in enumerate(doc_root.findall(".//w:p", namespaces=nsmap), 1):
        texts = []

        # Catch all descendant <w:r> elements
        run_elems = para.findall(".//w:r", namespaces=nsmap)

        if not run_elems:
            continue

        print(f"\nüìÑ Paragraph {para_idx}:")

        for run_idx, run in enumerate(run_elems, 1):
            t_elem = run.find(".//w:t", namespaces=nsmap)
            if t_elem is not None and t_elem.text:
                texts.append(t_elem.text)
                print(f"   üß© Run {run_idx}: \"{t_elem.text}\"")

        if not texts:
            print("   ‚ö†Ô∏è No text found in this paragraph's runs.")
            continue

        combined_text = ''.join(texts)
        print(f"   üîó Combined into one: \"{combined_text}\"")

        # üßπ Remove all runs ‚Äî now using XPath to target all
        for run in run_elems:
            try:
                run.getparent().remove(run)
            except ValueError:
                print("   ‚ùå Skipped a run that isn't a direct child.")

        # ‚ûï Add new combined run
        new_run = etree.Element(f"{{{nsmap['w']}}}r")
        new_text = etree.SubElement(new_run, f"{{{nsmap['w']}}}t")
        new_text.text = combined_text
        para.append(new_run)
        print("   ‚úÖ Replaced with a single run.")
```

```python
flatten_runs_to_single_run(doc_root, nsmap)
```

```python
def match_sentences_and_add_comments(doc_root, comments_root, sentence_list):
    comment_id = 0
    sentence_pointer = 0

    for para in doc_root.findall(".//w:p", namespaces=nsmap):
        run_map = []
        full_text = ""

        # Build full_text and map each run to its span
        for run in para.findall(".//w:r", namespaces=nsmap):
            t_elem = run.find(".//w:t", namespaces=nsmap)
            if t_elem is not None and t_elem.text:
                start = len(full_text)
                full_text += t_elem.text
                end = len(full_text)
                run_map.append((start, end, run))

        remaining_text = full_text
        local_pointer = 0

        while sentence_pointer < len(sentence_list):
            sentence = sentence_list[sentence_pointer].strip()
            if not sentence:
                sentence_pointer += 1
                continue

            rel_pos = remaining_text.find(sentence)
            if rel_pos == -1:
                break

            global_start = local_pointer + rel_pos
            global_end = global_start + len(sentence)

            # Find run that holds the sentence
            start_run = None
            end_run = None
            for start, end, run in run_map:
                if start <= global_start < end and start_run is None:
                    start_run = run
                if start < global_end <= end and end_run is None:
                    end_run = run
                if start_run and end_run:
                    break

            if start_run is None or end_run is None:
                print(f"‚ö†Ô∏è Couldn't find run for sentence: {sentence}")
                sentence_pointer += 1
                continue

            # Insert comment markers around run
            comment_start = etree.Element(f"{{{W_NS}}}commentRangeStart", {
                f"{{{W_NS}}}id": str(comment_id)
            })
            comment_end = etree.Element(f"{{{W_NS}}}commentRangeEnd", {
                f"{{{W_NS}}}id": str(comment_id)
            })
            comment_ref_run = etree.Element(f"{{{W_NS}}}r")
            etree.SubElement(comment_ref_run, f"{{{W_NS}}}commentReference", {
                f"{{{W_NS}}}id": str(comment_id)
            })

            start_run.addprevious(comment_start)
            end_run.addnext(comment_end)
            comment_end.addnext(comment_ref_run)

            # Add actual comment to comments.xml
            comment_elem = etree.SubElement(comments_root, f"{{{W_NS}}}comment", {
                f"{{{W_NS}}}id": str(comment_id),
                f"{{{W_NS}}}author": "AutoBot",
                f"{{{W_NS}}}date": "2025-06-01T00:00:00Z"
            })
            p = etree.SubElement(comment_elem, f"{{{W_NS}}}p")
            r = etree.SubElement(p, f"{{{W_NS}}}r")
            t = etree.SubElement(r, f"{{{W_NS}}}t")
            t.text = f"This is comment {comment_id + 1}"

            print(f"üí¨ Comment {comment_id + 1} added to: {sentence}")
            comment_id += 1
            sentence_pointer += 1
            local_pointer = global_end
            remaining_text = full_text[local_pointer:]
```

```python
def update_content_types(temp_dir):
    ct_path = os.path.join(temp_dir, "[Content_Types].xml")
    ct_tree = etree.parse(ct_path)
    ct_root = ct_tree.getroot()
    if not any(el.attrib.get("PartName") == "/word/comments.xml" for el in ct_root.findall(".//")):
        etree.SubElement(ct_root, "Override", {
            "PartName": "/word/comments.xml",
            "ContentType": "application/vnd.openxmlformats-officedocument.wordprocessingml.comments+xml"
        })
    ct_tree.write(ct_path, xml_declaration=True, encoding="utf-8")
```

```python
def update_relationships(temp_dir):
    rels_path = os.path.join(temp_dir, "word", "_rels", "document.xml.rels")
    rels_tree = etree.parse(rels_path)
    rels_root = rels_tree.getroot()
    if not any(rel.attrib.get("Target") == "comments.xml" for rel in rels_root):
        etree.SubElement(rels_root, "Relationship", {
            "Id": "rId100",
            "Type": "http://schemas.openxmlformats.org/officeDocument/2006/relationships/comments",
            "Target": "comments.xml"
        })
    rels_tree.write(rels_path, encoding="utf-8", xml_declaration=True)
```

```python
def rezip_docx(temp_dir, output_path):
    with zipfile.ZipFile(output_path, 'w', zipfile.ZIP_DEFLATED) as docx:
        for foldername, _, filenames in os.walk(temp_dir):
            for filename in filenames:
                filepath = os.path.join(foldername, filename)
                arcname = os.path.relpath(filepath, temp_dir)
                docx.write(filepath, arcname)
```

```python
doc_tree.write(doc_path, encoding="utf-8", xml_declaration=True)
etree.ElementTree(comments_root).write(comments_path, encoding="utf-8", xml_declaration=True)
update_content_types(temp_dir)
update_relationships(temp_dir)
rezip_docx(temp_dir, output_path)
shutil.rmtree(temp_dir)
```

```python
def insert_comment(doc_root, sentence, W_NS, nsmap, comment_id, comments_root):
    import re
    from lxml import etree

    normalized_target = re.sub(r'\s+', ' ', sentence.strip()).lower()
    
    for para in doc_root.findall(".//w:p", namespaces=nsmap):
        run_text = ""
        run_map = []
        
        # STEP 1: Gather run texts and track their positions
        for run in para.findall(".//w:r", namespaces=nsmap):
            t_elem = run.find(".//w:t", namespaces=nsmap)
            if t_elem is not None and t_elem.text:
                start = len(run_text)
                run_text += t_elem.text
                end = len(run_text)
                run_map.append((start, end, run))
                print(f"Run text: {t_elem.text}")
        
        print(f"\nFull paragraph combined: {run_text}")
        
        # Normalize both the paragraph and the target sentence
        normalized_para = re.sub(r'\s+', ' ', run_text.strip()).lower()
        normalized_target = re.sub(r'\s+', ' ', sentence.strip()).lower()
        
        print(f"Looking for: {normalized_target}")
        print(f"Normalized paragraph: {normalized_para}")
        
        # STEP 2: Find where the sentence starts and ends in the paragraph
        start_idx = normalized_para.find(normalized_target)
        if start_idx == -1:
            print("Sentence not found.")
            continue
        
        end_idx = start_idx + len(normalized_target)
        matched_substring = normalized_para[start_idx:end_idx]
        print(f"Match found: {matched_substring}")
        
        # STEP 3: Find runs that contain the start and end
        start_run = None
        end_run = None
        for start, end, run in run_map:
            if start <= start_idx < end and start_run is None:
                start_run = run
                print("Start run identified.")
            if start < end_idx <= end and end_run is None:
                end_run = run
                print("End run identified.")
            if start_run is not None and end_run is not None:
                break
        
        if start_run is None or end_run is None:
            print("Couldn't find exact run boundaries. Skipping comment.")
            continue
        
        #(Insert comment code would go here...)
        
        if start_run is not None and end_run is not None:
            comment_start = etree.Element(f"{{{W_NS}}}commentRangeStart", {f"{{{W_NS}}}id": str(comment_id)})
            comment_end = etree.Element(f"{{{W_NS}}}commentRangeEnd", {f"{{{W_NS}}}id": str(comment_id)})
            comment_ref_run = etree.Element(f"{{{W_NS}}}r")
            rpr = etree.SubElement(comment_ref_run, f"{{{W_NS}}}rPr")
            etree.SubElement(rpr, f"{{{W_NS}}}rStyle", {f"{{{W_NS}}}val": "CommentReference"})
            etree.SubElement(comment_ref_run, f"{{{W_NS}}}commentReference", {f"{{{W_NS}}}id": str(comment_id)})
        
            start_run.addprevious(comment_start)
            end_run.addnext(comment_end)
            end_run.addnext(comment_ref_run)
        
            comment = etree.SubElement(comments_root, f"{{{W_NS}}}comment", {
                f"{{{W_NS}}}author": "GPT",
                f"{{{W_NS}}}initials": "AI",
                f"{{{W_NS}}}date": "2025-05-29T00:00:00Z",
                f"{{{W_NS}}}id": str(comment_id)
            })
            p = etree.SubElement(comment, f"{{{W_NS}}}p")
            r = etree.SubElement(p, f"{{{W_NS}}}r")
            t = etree.SubElement(r, f"{{{W_NS}}}t")
            t.text = f"This is comment {comment_id + 1}"
            return True
            
    return False
```
